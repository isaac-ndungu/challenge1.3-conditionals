## 1

```python

import random

# generate random number between 0 and 100
target = random.randint(0, 100)
# print('Number to guess: ', target)

attempts = 1
while attempts <= 4:
    # guess from user
    try:
        guess = int(input('Guess a number between 0 and 100: '))
        if guess < 0 or guess > 100:
            raise Exception('Please enter an number between 0 and 100')
    except Exception as e:
        print(e)
        continue

    if guess < target:
        print('Wrong! The target number is higher, try again:')
        attempts += 1
    elif guess > target:
        print('Wrong! The target number is lower, try again:')
        attempts += 1
    else:
        print('Correct! You guessed it.')
        break
else: 
    print(f'Out of attempts! The number was {target}')




```


## 2

```python
first_num = int(input('Enter your first number: '))
operator = input('Enter operation to perform (+, -, *, /)').strip()
second_num = int(input('Enter your second number: '))


if operator == '+':
    result = first_num + second_num
    print(f'{first_num} + {second_num}')
    print('Answer: ', result)
elif operator == '-':
    result = first_num - second_num
    print(f'{first_num} - {second_num}')
    print('Answer: ', result)
elif operator == '*':
    result = first_num * second_num
    print(f'{first_num} * {second_num}')
    print('Answer: ', result)
elif operator == '/':
    result = first_num / second_num
    print(f'{first_num} / {second_num}')
    print('Answer: ', result)
else:
    print('enter a valid operator(+, -, *, /)')

```

## 3


```python
number = int(input('Enter a number to check if it is a prime number'))

i = 2
is_prime = True
while i <= int(number**0.5):
    if number % i == 0:
        is_prime = False
        break
    i += 1


if is_prime:
    print(f'{number} is a prime number')
else:
    print(f'{number} is not a prime number')
```
